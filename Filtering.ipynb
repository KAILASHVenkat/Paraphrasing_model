```python
import pandas as pd
url = "https://github.com/SoumyaSJha/TextToneTune/raw/main/dataset/t3_data%20copy.csv"
df = pd.read_csv(url)
df
```

```python
df[['e1', 'e2']] = df['input_text'].str.split(expand=True)[[0, 2]]
df['e2'] = df['e2'].str.strip(':')
df
```

```python
unique_e1_values = df['e1'].unique()
unique_e2_values = df['e2'].unique()

print("Unique values in 'e1' column:", unique_e1_values)
print("Unique values in 'e2' column:", unique_e2_values)
```

```python
high_negative_emotions = ['anger', 'disgust', 'grief', 'fear', 'sadness']
low_negative_emotions = ['nervousness', 'annoyance', 'disappointment', 'embarrassment', 'remorse', 'disapproval']
neutral_emotions = ['confusion', 'curiosity', 'realization', 'surprise']
low_positive_emotions = ['approval', 'caring', 'desire', 'relief']
high_positive_emotions = ['amusement', 'excitement', 'pride', 'optimism', 'gratitude', 'joy', 'admiration', 'love']

# Applying rules to filter DataFrame
filtered_df = df[
    (df['e1'].isin(high_positive_emotions) & ((df['e2'].isin(low_positive_emotions)) | (df['e2'].isin(neutral_emotions)))) |
    ((df['e1'].isin(low_positive_emotions) | df['e1'].isin(neutral_emotions)) & (df['e2'].isin(neutral_emotions))) |
    ((df['e1'].isin(low_negative_emotions)) & (df['e2'].isin(neutral_emotions))) |
    (df['e1'].isin(high_negative_emotions) & ((df['e2'].isin(low_negative_emotions)) | df['e2'].isin(neutral_emotions)))
]

# Displaying the filtered DataFrame
df = filtered_df.drop(['e1', 'e2'], axis=1)


df
```

```python
df.to_csv('filtered_data.csv', index=False)
```
